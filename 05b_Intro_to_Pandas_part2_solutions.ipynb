 # Exercise 1

In the SNPs datase, subset the dataset based on the most commont variant consequence, and take the SNP with the last (biggest) Chromosome/scaffold position end (bp)

```python
# Subset the dataset based on counts
# in this example, take only the SNPs of the most common variant type
counts = SNPs['Variant consequence'].value_counts()
#print(counts.index[0])
counts.index[0]
```

```python
# subset the dataset based on the most common variant type
subset = SNPs[SNPs['Variant consequence'] == counts.index[0]]
SNPs[SNPs['Variant consequence'] == counts.index[0]]
```

```python
# take the maximum
subset.sort_values(by ='Chromosome/scaffold position end (bp)', ascending= False)
```

```python
subset.sort_values(by ='Chromosome/scaffold position end (bp)', ascending= False).head(1)['Variant name']
```

```python
new_sorted_df = SNPs[SNPs['Variant consequence'] == counts.index[0]].sort_values(by ='Chromosome/scaffold position end (bp)', ascending= False)

new_sorted_df.reset_index(inplace=True)

new_sorted_df.loc[0,'Variant name']
```

# Exercise 2
What is the most ordered  item in the chipotle dataframe?
And how many of these items were ordered?
What is the cumulative revenue from this item?

```python
chipotle = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep= '\t')
chipotle.head()
```

```python
# The most ordered item in the item_name column:
chipotle['item_name'].value_counts()
```

```python
# The most ordered item is:
chipotle['item_name'].value_counts().index[0]
```

```python
# From the most ordered items, there were ordered a total of # items:

most_ordered = chipotle.groupby(['item_name'])['quantity'].sum().sort_values(ascending= False).index[0]
chipotle.groupby(['item_name'])['quantity'].sum().sort_values(ascending= False)[0]
```

```python
chipotle.groupby(['item_name'])['quantity'].sum()
```

```python
chipotle.groupby(['item_name'])['quantity'].mean()
```

```python
chipotle
```

```python
chipotle.groupby(['item_name']).sum().sort_values(by ='quantity', ascending= False).head(1)
```

```python
# The cumulative price was:
# modify price to actually be a float
chipotle['item_price'] = chipotle['item_price'].str.split('$').str[1].astype(float)
```

```python
chipotle.groupby(by = 'item_name').sum().loc[most_ordered, 'item_price']
```

# Exercise 3

Show the following individual dataframes then merge them



```python
import pandas as pd
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})

```

Show the resulting dataframe, then merge it with the following:

```python
df1_2 = pd.merge(df1, df2,  on ='employee', how='outer')
```

```python
df1_2
```

```python
df3 = pd.DataFrame({'group': ['Accounting', 'Engineering'],
                    'supervisor': ['Carly', 'Guido']})
```

```python
df1_2_3 = pd.merge(df1_2, df3,  on = 'group')
df1_2_3
```

Try the inner and outer method, and the left or right method.

```python
pd.merge(df1_2, df3,  on = 'group', how='outer')
```

```python
pd.merge(df1_2, df3,  on = 'group', how='inner')
```

```python
pd.merge(df1_2, df3,  on = 'group', how='left')
```

```python
pd.merge(df1_2, df3,  on = 'group', how='right')
```
